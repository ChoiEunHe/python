# 웹상의 정보를 추출하는 방법 
- urllib.request 모듈은 웹사이트에 있는 데이터에 접근하는 기능을 제공

```python
# 라이브러리 불러오기 
import urllib.request as req

# 다운로드 파일 이름 결정 
fileName = "mov01.jpg"

# URL 과 저장 경로 지정하기 
url = "https://zeushahn.github.io/Test/images/" + fileName
saveName = "./Data/" + fileName

# 다운로드 
req.urlretrieve(url, saveName)
print('OK')
```

```python
### Exercise
## https://zeushahn.github.io/Test/images/ 에서 mov02.jpg, mov03.jpg, mov04.jpg

# 라이브러리 불러오기 
# import urllib.request as request

# 다운로드 파일 이름 결정 
# fileLists = [ 'mov02.jpg', 'mov03.jpg', 'mov04.jpg', 'mov05.jpg','mov06.jpg','mov07.jpg', 'mov08.jpg' ]
import urllib.request as req

fileLists = [f"mov{i}.jpg" for i in range(2, 9)]

for item in fileLists:
    url = "https://zeushahn.github.io/Test/images/" + item
    saveName = "./Data/" + item
    req.urlretrieve(url, saveName)
    print(item, "is downloaded")
```

```python
# mov04.jpg

import urllib.request as req

# 다움로든 파일 이름 결정 
fileName = "mov05.jpg"
url = f"https://zeushahn.github.io/Test/images/{fileName}"
print(url)
saveName = f"./Data/{fileName}"

# 다운로드
mem = req.urlopen(url).read()

# 파일 저장하기 
with open(saveName, mode="wb") as f:
  f.write(mem)
  print('OK!')


```

```python
# 기상청 RSS 
# https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109

# 지역코드 > 전국 : 108,서울/경기도 : 109,강원도 : 105

import urllib.request as req
# 지역코드 
stnId= '109'

# 기상청 RSS 사이트 주소 
url = f"http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId={stnId}"
print(url)

data = req.urlopen(url).read().decode('utf-8')
print(data)
```
