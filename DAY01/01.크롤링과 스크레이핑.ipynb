# 크롤링과 스크래이핑
## requests
응답코드 
1xx : 요청은 받았고 작업 진행 중        
2xx : 사용자의 요청이 성공적으로 수행됨.   
3xx : 요청은 완료되었으나 리다이렛션이 필요.   
4xx : 사용자의 요청이 잘못됨.   
5xx : 서버에 오류가 발생됨.   


```python
# request 응답코드
import requests as req

URL = 'https://www.google.com'
response = req.get(URL)
if (response.status_code == 200) :
   print("연결이 잘 되었음")
else :
   print("연결에 문제가 발생함")
```

```python
# Naver의 메인페이지 정보 가져오기
import requests as req

URL = 'https://www.naver.com'
response = req.get(URL)

print(response.text)
```

```python
# 네이버에서 정보 검색 결과 가져오기 
import requests as req

URL = 'https://search.naver.com/search.naver'
params = {'query' : 'python'}
response = req.get(URL,params=params)

print(response.text)
```

```python
# 네이버에서 정보 검색 결과 가져오기 
import requests as req

URL = 'https://comic.naver.com/webtoon/detail.nhn'
params = {'titleId' : '687915', 'no' : '157'}
response = req.get(URL,params=params)

print(response.text)
```

```python
# JSON 형식의 데이타 가져오기 

import requests as req
response = req.get("https://raw.githubusercontent.com/naver/naver-openapi-guide/draft/naver-openapi-swagger.json")

result = response.json()
print(type(result))
print(result)
```
